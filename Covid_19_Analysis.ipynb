# Data Cleaning

```python
# check the missing values
df.count()
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull())
plt.show()
```

Show the number of confirmed, Deaths and Recovered Cases in each Region

```python
df.head(5)
```

```python
# Region with most confirmed cases
df.groupby('Region')['Confirmed'].sum().sort_values(ascending=False)
```

```python
# Removing records where all the cases are less than 10
df=df[~(df.Confirmed < 10)]
```

```python
df.head()
```

# Exploratory Data Analysis

```python
# Region with minimum number of deaths
df.groupby('Region').Deaths.sum().sort_values(ascending=True).head(50)
```

```python
df[df.Region == 'India']
```

```python
# sorting data with respect to confirmed cases in ascending order
df.sort_values(by=['Recovered'])
```

```python
# Recovered cases
df.sort_values(by=['Recovered'],ascending = False ).head(50)
```
